```python
Shape = [15,15]
```

```python
A = np.random.rand(Shape[0],Shape[1])
```

```python
A_copy = A.copy()
Q = np.eye(Shape[0])
Image_Files = []
ctr = 0
for j in range(Shape[1]):
    tau = (A_copy[j:,j]@A_copy[j:,j])**0.5
    # print(tau)
    sign = np.sign(A_copy[j,j])
    y = np.zeros([Shape[0]-j,1])
    y[0,0] = -1*sign*tau
    # print(y)
    x = A_copy[j:,j]
    x = x.reshape([Shape[0]-j,1])
    u_hat = x - y
    # print(x)
    u = u_hat/((u_hat.T@u_hat)**0.5)
    # print(u)
    I_1 = np.eye(Shape[0])
    I = np.eye(Shape[0]-j)
    Q_temp = I - (2*u@u.T)
    I_1[j:,j:] = Q_temp
    Q_temp = I_1
    Q = Q@Q_temp
    A_copy = Q_temp.T@A_copy
    # print(A_copy)
    # sns.heatmap(A_copy,cmap="viridis",vmin=Min,vmax=Max)
    # plt.savefig(f"Plot_{ctr}.png")
    # Image_Files.append(f"Plot_{ctr}.png")
    # ctr = ctr + 1
    # plt.show()
```

```python
A_copy = A.copy()
Q = np.eye(Shape[0])
Image_Files = []
ctr = 0
for j in range(Shape[1]):
    tau = (A_copy[j:,j]@A_copy[j:,j])**0.5
    # print(tau)
    sign = np.sign(A_copy[j,j])
    y = np.zeros([Shape[0]-j,1])
    y[0,0] = -1*sign*tau
    # print(y)
    x = A_copy[j:,j]
    x = x.reshape([Shape[0]-j,1])
    u_hat = x - y
    # print(x)
    u = u_hat/((u_hat.T@u_hat)**0.5)
    # print(u)
    I_1 = np.eye(Shape[0])
    I = np.eye(Shape[0]-j)
    Q_temp = I - (2*u@u.T)
    I_1[j:,j:] = Q_temp
    Q_temp = I_1
    Q = Q@Q_temp
    A_copy = Q_temp.T@A_copy
    # print(A_copy)
    sns.heatmap(A_copy,cmap="viridis",vmin=Min,vmax=Max)
    plt.savefig(f"Plot_{ctr}.png")
    Image_Files.append(f"Plot_{ctr}.png")
    ctr = ctr + 1
    plt.show()
```
